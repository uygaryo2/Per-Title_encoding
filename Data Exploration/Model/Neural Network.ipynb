```python
import pandas as pd
from sklearn.svm import SVR
from joblib import load, dump
from sklearn.metrics import mean_squared_error as mse

from sklearn.linear_model import LinearRegression
```

```python
train_data = load('data/train.df')
X_train = train_data.drop(columns=['vmaf'])
y_train = train_data['vmaf']
```

```python
svr_model = SVR(kernel='linear')
svr_model.fit(X_train, y_train)
y_pred = svr_model.predict(X_train)
print(mse(y_pred, y_train))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred, y_train)
plt.show()
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_train)
print(mse(y_pred_lr, y_train))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred_lr, y_train)
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor

nn_model = MLPRegressor(hidden_layer_sizes=(100, 250, 50))
nn_model.fit(X_train, y_train)
y_pred_nn = nn_model.predict(X_train)
print(mse(y_pred_nn, y_train))
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_pred_nn, y_train)
plt.show()
```

```python
train_data = load('data/test.df')
X_test = train_data.drop(columns=['vmaf'])
y_test = train_data['vmaf']

y_pred_nn = nn_model.predict(X_test)
print(mse(y_pred_nn, y_test))

import matplotlib.pyplot as plt

plt.scatter(y_pred_nn, y_test)
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_train)
print(mse(y_pred_rf, y_train))

import matplotlib.pyplot as plt

plt.scatter(y_pred_rf, y_train)
plt.show()
```
