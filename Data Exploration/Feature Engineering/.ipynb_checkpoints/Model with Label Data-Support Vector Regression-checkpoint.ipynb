### normalise numerical data

```python
from sklearn.preprocessing import MinMaxScaler
from joblib import dump, load
import numpy as np

numerical_columns =['clip_duration', 'clip_size', 'clip_bitrate_total', 'encode_bitrate_video','crf', 'vmaf'] 

for column in numerical_columns:
    normalizer = MinMaxScaler()
    normalizer.fit(train[column].values.reshape(-1,1))
    dump(normalizer, f"{column}_normalizer.pkl")
    train[column] = normalizer.transform(train[column].values.reshape(-1,1))
    test[column] = normalizer.transform(test[column].values.reshape(-1,1))
    
```

```python
train.shape
```

### Test Model

```python
y_train = train['vmaf']
X_train = train.drop(columns=['vmaf'])

y_test = test['vmaf']
X_test = test.drop(columns=['vmaf'])
```

```python
from sklearn.linear_model import LinearRegression

linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train)
print(linear_regression_model.coef_)
```

```python
from sklearn.metrics import mean_absolute_error as mae
from yellowbrick.regressor import PredictionError

y_pred = linear_regression_model.predict(X_train)
y_scaler = load('vmaf_normalizer.pkl')
y_pred = y_scaler.inverse_transform(np.array(y_pred).reshape(-1,1))
y_train = y_scaler.inverse_transform(np.array(y_train).reshape(-1,1))

print("Linear Regression - Mean Absolute Error:", mae(y_pred=y_pred, y_true=y_train))
print("Linear Regression - Mean Absolute Percentage Error:", np.mean(abs(y_train-y_pred)/abs(y_train)),"%")
```

```python
from sklearn.metrics import mean_absolute_error as mae

y_pred_test = linear_regression_model.predict(X_test)
y_scaler = load('vmaf_normalizer.pkl')
y_pred_test = y_scaler.inverse_transform(np.array(y_pred_test).reshape(-1,1))
y_test = y_scaler.inverse_transform(np.array(y_test).reshape(-1,1))

print("Linear Regression - Mean Absolute Error:", mae(y_pred=y_pred_test, y_true=y_test))
print("Linear Regression - Mean Absolute Percentage Error:", np.mean(abs(y_test-y_pred_test)/abs(y_test)),"%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(abs(y_test-y_pred_test)/abs(y_test))
plt.show()
```
